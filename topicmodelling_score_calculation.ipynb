```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("depressed_users_cleaned.csv",encoding='latin1')
df
```

```python
df.dropna(inplace=True)
df
```

## Heartbreak

```python
heartbreak = 'feel everyth empti better anymor hurt doesnt noth sick away,like look sometim actual day treat tweet would theyr liter,dont anymor even understand anyth care mind mean live give,cant wait stop even believ tell anyth stand anymor thought,peopl mental ill health someon care abus beauti twitter tweet'
```

```python
Name_list = df["tweet"].tolist()
Name_list
```

```python
vectorizer = TfidfVectorizer()
# To make uniformed vectors, both documents need to be combined first.
Name_list.append(heartbreak)
embeddings = vectorizer.fit_transform(Name_list)
heartbreak_score = cosine_similarity(embeddings[-1:], embeddings[0:-1]).flatten()
```

```python
embeddings[0:-1]
```

```python
heartbreak_score.tolist()
```

```python
heartbreak_score.shape
```

```python
df['heartbreak']=heartbreak_score
```
